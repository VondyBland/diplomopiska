```python
import requests


def get_route_info_here(origin, destination, api_key, mode="fastest;car"):
    url = f"https://route.ls.hereapi.com/routing/7.2/calculateroute.json?apiKey={api_key}&waypoint0={origin}&waypoint1={destination}&mode={mode}"
    response = requests.get(url)
    print(response)
    data = response.json()
    distance = data["response"]["route"][0]["summary"]["distance"]
    shape = data["response"]["route"][0]["shape"]
    route_info = {
        "distance": data["response"]["route"][0]["summary"]["distance"],
        "duration": data["response"]["route"][0]["summary"]["baseTime"]
    }
    return route_info,shape,distance

# Пример использования функций
api_key_here = "MV5O01Bda5AC0xMXhDX2xuf4qH0lMEESAlhwA3qV-dk"
origin_here = "60.066913,30.334023"  # Координаты начальной точки (широта, долгота)
destination_here = "60.051176,30.332676"  # Координаты конечной точки (широта, долгота)

# Для машины
route_info_here_car = get_route_info_here(origin_here, destination_here, api_key_here, mode="fastest;car")


print("Затраченное время для машины:", route_info_here_car)
```
